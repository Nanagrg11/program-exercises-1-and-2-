```python
#question3
student= int(input("how many student?: "))
group_size= int(input("required group size?: "))
full_group=student//group_size
left_group=student%group_size
if full_group ==1 and left_group !=1:
    print(f"there will be {full_group} groups with {left_group} students left over")
elif full_group !=1 and left_group ==1: 
    print(f"there will be {full_group} groups with {left_group} students left over")
elif full_group ==1 and left_group ==1:
    print(f"there will be {full_group} groups with {left_group} students left over")
else:
    print(f"there will be {full_group} groups with {left_group} students left over")
    
 
```
